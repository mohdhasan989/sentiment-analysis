```python
df=pd.read_csv("py.csv",encoding='ISO-8859-1')
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
import re
from nltk.corpus import stopwords
```

```python

texts = df['text']
sentiment = df['sentiments']
df.head
```

```python
def preprocess_text(text):
    # Tokenize the text
    tokens = text.lower() # Convert to lower case and tokenize
    tokens = re.sub(r'[^a-zA-Z\s]', '', text)
    tokens = text.split()
       # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words and word.isalpha()]  # Only keep alphabetic words
    return " ".join(tokens)
    from nltk.stem import WordNetLemmatizer
    from nltk.tokenize import word_tokenize
    from nltk.stem import PorterStemmer
  
    lemmatizer = WordNetLemmatizer()
    stemmer = PorterStemmer()
        # Apply Stemming
    tokens = [stemmer.stem(word) for word in tokens]
    return " ".join(tokens)
text = texts.apply(preprocess_text)
print(text)
      
   





 
```

```python
sentiments = sentiment.map({'negative': 0, 'positive': 1})
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(texts, sentiment, test_size=0.2, random_state=42)
y_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Convert the text data into numerical features using CountVectorizer
vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
# Train a Naive Bayes model
model = MultinomialNB()
model.fit(X_train_vectorized, y_train)

```

```python
# Make predictions on the test set
y_pred = model.predict(X_test_vectorized)
print(y_pred)

```

```python
from sklearn.metrics import accuracy_score, classification_report
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')
print('Classification Report:')
print(classification_report(y_test, y_pred,))
```
